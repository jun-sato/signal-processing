```python
wav,sr = librosa.load('sample.wav')
```

```python
mel = librosa.feature.melspectrogram(wav,sr=sr)
mfcc = librosa.feature.mfcc(wav,sr=sr)
cqt = librosa.feature.chroma_cqt(wav,sr=sr)#
chroma = librosa.feature.chroma_cens(wav,sr=sr)
S, phase = librosa.magphase(librosa.stft(wav))
rms = librosa.feature.rmse(S=S)
S = np.abs(librosa.stft(wav))
contrast = librosa.feature.spectral_contrast(S=S, sr=sr)
```

```python
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(mel,
                                             ref=np.max),
                         y_axis='mel', fmax=8000,
                         x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(cqt, y_axis='chroma', x_axis='time')
plt.title('chroma_cqt')
plt.colorbar()
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.title('chroma_cens')
plt.colorbar()
plt.tight_layout()
```
